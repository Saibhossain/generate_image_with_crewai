```python
from crewai import LLM

from google.colab import userdata

gemini_api_key = userdata.get('google_api')

llm = LLM(
    model="gemini/gemini-2.5-flash",
    api_key= gemini_api_key
)
```

```python
from crewai import Agent, Task, Crew

researcher = Agent(
    role="Researcher",
    goal="Find AI trends",
    backstory="Expert analyst",
    llm=llm
)
```

```python
researcher
```

```python
writer = Agent(
    role="Writer",
    goal="Write article",
    backstory="Creative writer",
    llm=llm
)

# Tasks
task1 = Task(
    description="Research AI trends",
    expected_output="A comprehensive report on the latest AI trends, including key developments, challenges, and future predictions.",
    agent=researcher
)

task2 = Task(
    description="Write article",
    expected_output="A well-written, engaging article, approximately 800 words, based on the research findings about AI trends.",
    agent=writer
)

# Crew
crew = Crew(
    agents=[researcher, writer],
    tasks=[task1, task2]
)

# Run
result = crew.kickoff()
print(result)
```
